```python
import pandas as pd
import duckdb

# 1. قراءة ملف الإكسيل الخاص بك
df = pd.read_excel("decodelabes-internship.xlsx")

# 2. ربط الكشكول بـ SQL وتشغيل قاعدة البيانات في الذاكرة
%load_ext sql
%sql duckdb:///:memory:

# 3. تسجيل الـ DataFrame داخل DuckDB لكي يراها كجدول اسمه orders
duckdb.register("orders", df)
```

```python
import pandas as pd
import duckdb

# 1. قراءة الملف من جديد وضمان وجوده في الذاكرة
df = pd.read_excel("decodelabes-internship.xlsx")
duckdb.register("orders", df)

# 2. تشغيل استعلام الـ SQL مباشرة وعرض النتيجة فوراً
result = duckdb.query("""
    SELECT 
        COUNT(OrderID) AS total_orders, 
        SUM(TotalPrice) AS total_revenue
    FROM orders
""").df()

# عرض النتيجة على الشاشة
result
```
